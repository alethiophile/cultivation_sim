## Presets

```python
minor_alchemy = {
    'alchemy_minor': True,
    'pill_overflow': lambda: False
}

major_alchemy = {
    'alchemy_minor': True,
    'alchemy_major': True,
    'pill_overflow': lambda: False
}

pills_last = {
    'willpower_pills_ctr': 6,
    'willpower_pills_policy': 'last'
}

pills_first = {
    'willpower_pills_ctr': 6,
    'willpower_pills_policy': 'first'
}

pills_after_fail = {
    'willpower_pills_ctr': 6,
    'willpower_pills_policy': 'after-fail'
}

def odds(a):
    if a.day == 1:
        a.pill_cooldown = 5
        a.remove_nopill_rp = False
        a.pill_overflow = lambda: False
    elif a.attachments == 7:
        a.pill_cooldown = 4
```

## Alchemical Boosting option

These runs use the `minor_alchemy` preset, corresponding to Alchemical Boosting. They assume a will modifier for Sefwen of 13 (best guess at current value) or 14 (for if we can give her a +1 bonus, e.g. via playing to her submission reflex).

```python
print_runs_data(run_many_sims(data={**minor_alchemy, 'sefwen_will_mod': 13}))
```

```python
print_runs_data(run_many_sims(data={**minor_alchemy, 'sefwen_will_mod': 14}))
```

```python
print_runs_data(run_many_sims(data={**minor_alchemy,
                                    **pills_last,
                                    'sefwen_will_mod': 14, 
                                    }))
```

```python
print_runs_data(run_many_sims(data={**minor_alchemy,
                                    **pills_first,
                                    'sefwen_will_mod': 14, 
                                    }))
```

```python
print_runs_data(run_many_sims(data={**minor_alchemy,
                                    **pills_after_fail,
                                    'sefwen_will_mod': 14, 
                                    }))
```

## Playing the Odds option

These runs use Playing the Odds. The function assumes we'll eventually get 4 days between removals.

```python
print_runs_data(run_many_sims(timed_adjust=odds, data={'sefwen_will_mod': 13}))
```

```python
print_runs_data(run_many_sims(timed_adjust=odds, data={'sefwen_will_mod': 14}))
```

```python
print_runs_data(run_many_sims(timed_adjust=odds, 
                              data={'sefwen_will_mod': 14,
                                   **pills_after_fail}))
```

```python
print_runs_data(run_many_sims(timed_adjust=odds, 
                              data={'sefwen_will_mod': 14,
                                   **pills_after_fail,
                                   'failure_insurance_ctr': 4}))
```
